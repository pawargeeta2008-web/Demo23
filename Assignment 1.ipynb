```python
# 1. Take two numbers as input from the user and print their Sum , Difference, Product and Quotient.
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Sum =", num1 + num2)
print("Difference =", num1 - num2)
print("Product =", num1 * num2)

if num2 != 0:
    print("Quotient =", num1 / num2)
else:
    print("Cannot divide by zero")
```

```python
# 2. Accept the temperature in Celcius from the user convert in to Fehrenheit and Kelvin and print all three values with lables.
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15

print("Celsius =", celsius)
print("Fahrenheit =", fahrenheit)
print("Kelvin =", kelvin)
```

```python
# 3. Take a string as a input and print its uppercase version, its length and the string reversed
text = input("Enter a string: ")

print("Uppercase:", text.upper())
print("Length:", len(text))
print("Reversed:", text[::-1])
```

```python
# 4. Take three Numbers as a input and print the largest among them. Do not use built in max() function.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

largest = a

if b > largest:
    largest = b

if c > largest:
    largest = c

print("Largest number =", largest)
```

```python
# 5. Take a year as a input and print whether it is a leap year or not.
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
# 6. Build a simple calculater - Take two numbers and an operator (+,-,*,/) as input, perform the operation, and print the result. Print an error message for invalid operators or division by zero.
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+,-,*,/): ")

if op == "+":
    print("Result =", num1 + num2)
elif op == "-":
    print("Result =", num1 - num2)
elif op == "*":
    print("Result =", num1 * num2)
elif op == "/":
    if num2 != 0:
        print("Result =", num1 / num2)
    else:
        print("Division by zero is not allowed")
else:
    print("Invalid Operator")
```

```python
# 7. Take a number as a input and print its Multiplication Table from 1 to 10. Format each line as: 7*3 = 21
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
# 8. Take a positive integer n as input and print all prime numbers from 2 to n.
n = int(input("Enter n: "))

for num in range(2, n + 1):
    prime = True

    for i in range(2, num):
        if num % i == 0:
            prime = False
            break

    if prime:
        print(num)
```

```python
# 9. Keep accepting numbers from the user until they enter 0. Then print the total count of numbers entered, their sum and their Average. Do not include 0 in the calculation.
count = 0
total = 0

while True:
    num = int(input("Enter a number (0 to stop): "))

    if num == 0:
        break

    count += 1
    total += num

if count > 0:
    average = total / count
    print("Count =", count)
    print("Sum =", total)
    print("Average =", average)
else:
    print("No numbers entered")
```

```python
# 10. Take n as a input and print a triangle pattern of n rows made of asterisks, where row i has (2i-1) stars centered appropriately.
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```
